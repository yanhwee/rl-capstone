### Imports

```python
import gym
from sole import SOLE
import numpy as np
```

### Taxi

```python
taxi_env = gym.make('Taxi-v3')
taxi_sole = SOLE(taxi_env, gamma=0.9)
taxi_sole = SOLE(taxi_env, gamma=0.9, 
                 eval_q_table=taxi_sole.get_q_table())
```

```python
taxi_sole.test(delay=0.1)
```

```python
taxi_sole.stats('Taxi')
```

### Frozen Lake

```python
lake_env = gym.make('FrozenLake8x8-v0')
lake_sole = SOLE(lake_env, gamma=0.9)
lake_sole = SOLE(lake_env, gamma=0.9, 
                 eval_q_table=lake_sole.get_q_table())
```

```python
lake_sole.test(delay=0.1)
```
