### Imports

```python
import gym
from qlearning import QLearning
from dp import DP
from sole import SOLE
import numpy as np
```

### Taxi

```python
taxi_env = gym.make('Taxi-v3')
taxi_eval = DP(taxi_env, gamma=0.9, policy_stop=False)
```

```python
taxi_qlearning = QLearning(
    taxi_env, gamma=0.9,
    epsilon_start=0.5, epsilon_end=0.4, 
    alpha_start=0.4, alpha_end=0.2,
    max_eps=3000, n_step=1, eval_q_table=taxi_eval.get_q_table())
```

```python
taxi_qlearning.stats('Taxi-v3 - Q-Learning')
```
