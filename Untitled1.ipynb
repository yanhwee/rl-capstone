```python
import gym
from sarsa import Sarsa
from dp import DP
from sole import SOLE
import numpy as np
```

```python
taxi_env = gym.make('Taxi-v3')
taxi_sole = DP(taxi_env, gamma=0.9, limit=1)
taxi_eval = DP(taxi_env, gamma=0.9, policy_stop=False)
```

```python
taxi_sarsa = Sarsa(
    taxi_env, gamma=0.9,
    epsilon_start=0.1, epsilon_end=0.001, 
    alpha_start=0.35, alpha_end=0.2,
    max_eps=3000, n_step=1, 
    init_q_table=taxi_sole.get_q_table(),
    eval_q_table=taxi_eval.get_q_table())
taxi_sarsa.stats()
```
