### Imports

```python
import gym
from montecarlo import MonteCarlo
from dp.dp import DP
from sole.sole import SOLE
import numpy as np
```

### Taxi

```python
taxi_env = gym.make('Taxi-v3')
taxi_eval = DP(taxi_env, gamma=0.9, policy_stop=False)
```

```python
taxi_mc = MonteCarlo(
    taxi_env, gamma=0.9, 
    epsilon_start=0.7, epsilon_end=0.001, 
    alpha_start=0.4, alpha_end=0.2, 
    max_eps=30000, eval_q_table=taxi_eval.get_q_table())
```

```python
taxi_mc.stats('Taxi-v3 - Monte Carlo')
```

```python
taxi_mc = MonteCarlo(
    taxi_env, gamma=0.9, 
    epsilon_start=0.7, epsilon_end=0.000, 
    alpha_start=0.4, alpha_end=0.1, 
    max_eps=100000, eval_q_table=taxi_eval.get_q_table())
```

```python
taxi_mc.stats()
```

```python
taxi_mc.test(delay=0.1)
```
