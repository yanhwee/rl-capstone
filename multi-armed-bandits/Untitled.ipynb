## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## k-armed bandit problem

```python
n_actions = 10
q_sigma = 1
r_sigma = 1
seed = 4
plot_sample = 10000
np.random.seed(seed)
q_values = np.random.normal(0, q_sigma, n_actions)
```

To visualise the k-armed bandit problem, we will plot the reward distribution among all actions using the violen plot.

```python
data = [np.random.normal(q_value, r_sigma, plot_sample) for q_value in q_values]
plt.violinplot(data, showextrema=False, showmeans=True)
plt.title(f'The {n_actions}-armed bandit problem')
plt.xlabel('Action')
plt.ylabel('Reward Distribution')
plt.xticks(range(1, n_actions + 1))
plt.show()
```

This function will produce a sample reward given an action

```python
def sample_action_reward(action):
    return np.random.normal(q_values[action], r_sigma)
```

For example, this is a sample reward by taking the first action

```python
sample_action_reward(0)
```
