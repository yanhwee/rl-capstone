```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, cos, sin
from IPython.display import clear_output
from tqdm.auto import tqdm
```

```python
f = lambda x: x
xs = np.linspace(0, 1)
ys = np.vectorize(f)(xs)
```

```python
plt.plot(xs,ys)
```

```python
a = lambda k, T, f, n: 2 / T * np.mean([f(x) * cos(2 * pi * k * x / T) for x in np.linspace(0, T, n)])
b = lambda k, T, f, n: 2 / T * np.mean([f(x) * sin(2 * pi * k * x / T) for x in np.linspace(0, T, n)])
```

```python
h = lambda x, T, f, n, m: \
    a(0, T, f, n) / 2 + \
    sum(b(k, T, f, n) * sin(2 * pi * k * x / T) for k in np.arange(1, m + 1)) + \
    sum(a(k, T, f, n) * cos(2 * pi * k * x / T) for k in np.arange(1, m + 1))# + \
    
```

```python
T = 1
n = 100
m = 21
```

```python
zs = [h(x, T, f, n, m) for x in tqdm(xs)]
plt.plot(xs, zs)
plt.show()
```

```python
for mi in range(m):
    zs = [h(x, T, f, n, mi) for x in xs]
    clear_output(wait=True)
    plt.plot(xs, zs)
    plt.show()
```
