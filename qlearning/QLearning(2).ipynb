### Imports

```python
import gym
from qlearning import QLearning
from expectedsarsa.expectedsarsa import ExpectedSarsa
from dp.dp import DP
from sole.sole import SOLE
import numpy as np
```

### Frozen Lake

```python
taxi_env = gym.make('FrozenLake-v0')
taxi_eval = SOLE(taxi_env, gamma=0.9)
```

```python
taxi_qlearning = QLearning(
    taxi_env, gamma=0.9,
    epsilon_start=0.5, epsilon_end=0.2, 
    alpha_start=0.4, alpha_end=0.2,
    max_eps=3000, n_step=1, eval_q_table=taxi_eval.get_q_table())
taxi_qlearning.stats('FrozenLake-v0 - Q-Learning')
```

```python
taxi_expectedsarsa = ExpectedSarsa(
    taxi_env, gamma=0.9,
    epsilon_start=0.5, epsilon_end=0.2, 
    alpha_start=0.4, alpha_end=0.2,
    max_eps=3000, n_step=1, eval_q_table=taxi_eval.get_q_table())
taxi_expectedsarsa.stats('FrozenLake-v0 - ExpectedSarsa')
```
